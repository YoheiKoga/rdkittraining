```python
from rdkit import rdBase, Chem
from rdkit.Chem import AllChem, Draw, rdMolDescriptors
from rdkit.Chem.Draw import SimilarityMaps

import pubchempy as pcp
```

```python
tmp = pcp.get_compounds('Paclitaxel', 'name')
```

```python
Chem.MolFromSmiles(tmp[0].isomeric_smiles)
```

```python
len(tmp)
```

```python
mols = [x for x in tmp if x is not None]
len(mols)
```

```python
mols
```

```python
taxol_list = []
for mol in mols:
    taxol_list.append(Chem.MolFromSmiles(mol.canonical_smiles))
```
